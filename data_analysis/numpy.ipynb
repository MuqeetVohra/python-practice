NumPy - Numerical Python

Advantages of Numpy Arrays:
    1. Allows several Mathematical Operations
    2. Faster operations


```python
import numpy as np
```

List vs Numpy - Time Taken

```python
# Time taken by list
python_list = [i for i in range(100000000)]
%time python_list = [i + 5 for i in python_list]
```

```python
np_array = np.array([i for i in range(100000000)])
%time np_array += 5
```

```python
# list
list1 = [1,2,3,4,5]
print(list1)
type(list1)
```

```python
# 1d array
np_array = np.array([1,2,3,4,5])
print(np_array)
type(np_array)
```

```python
np_array.shape
```

```python
# 2d array
np_array2 = np.array([(1,2,3,4),
                      (1,2,3,4)])
np_array2.shape
```

```python
# floating point values in array
np_array3 = np.array([1.1, 23.3, 21.3],dtype=float)
print(np_array3)
type(np_array3)
```

```python
# initial placeholders in numpy arrays

# create numpy array of 0
np_array4 = np.zeros((2,6)) # in () shape is mentioned
print(np_array4)

# create a numpy array of ones
np_array5 = np.ones((6,2))
print(np_array5)

np_array6 = np.full((5,3),5)
print(np_array6)
```

```python
# create an identity matrix
np_array7 = np.eye(4) 
np_array7
```
