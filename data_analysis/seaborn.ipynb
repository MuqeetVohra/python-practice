```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Importing dataset
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
# visualize the data
sns.relplot(data=tips_df, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

```python
# load the iris dataset
iris_df = sns.load_dataset('iris')
iris_df.head()
```

```python
# setting up theme for plotting
sns.set_theme('notebook')
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris_df)
```

```python
boston_df = pd.read_csv('data\\boston.csv')
```

```python
boston_df
```

```python
boston_df['PRICE'] = np.random.randint(10, 50, 506)
```

```python
boston_df.head()
```

```python
# distribution plot
sns.displot(x='PRICE', data=boston_df)
```

```python
correlation = boston_df.corr()
```

```python
# plotting Heat Map
plt.figure(figsize=(8, 8))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size': 8}, cmap='Blues')
```
