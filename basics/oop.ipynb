```python
# Constructor (__init__)
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
```

```python
# Creating an object
c = Car('Toyota', 'Corolla')
print(c.brand)
print(c.model)
```

```python
# Functions inside a Class
class Car:
    def __init__(self, brand):
        self.brand = brand
    
    def start(self):
        return f'{self.brand} is starting...'
```

```python
# using method
car = Car('Honda')
car.start()
```

```python
# Inheritance
# one class taking features from another
class Animal:
    def speak(self):
        return 'Animal speaks'

class Dog(Animal):
    pass

#created object
d = Dog()
d.speak()
```

```python
# Method overriding
class Animal:
    def speak(self):
        return 'Animal sound'
# Dog is a child class of animal    
class Dog(Animal):
    def speak(self):
        return 'Bark!'
    
d = Dog()
d.speak()
```

```python
class Bank:
    def __init__(self):
        self._balance = 1000 # '_' for Protected
    def deposit(self, amount):
        self._balance += amount
    # getter
    def get_balance(self):
        return self._balance
    # setter
    def set_balance(self, amount):
        if amount > 0:
            self._balance = amount
        else:
            return f'Invalid amount! Balance cannot be negative.'

b = Bank()
b.deposit(30)
b.set_balance(23)
b.get_balance()
```

```python
class Bank:
    def __init__(self):
        self.__balance = 2000 # private attribute
    # getter
    def get_balance(self):
        return self.__balance
    # setter
    def deposit(self, amount):
        self.__balance += amount
b = Bank()
# setting up value using setter method
b.deposit(230)
# getting up value using getter method
b.get_balance()
#b.__balance it will give error while accessing balance directly 
```

```python
# Polymorphism 
# same method name, different behavior
class Cat:
    def sound(self):
        return 'Meow'
class Dog:
    def sound(self):
        return 'Bark'
    
c = Cat()
d = Dog()
print(c.sound())
print(d.sound())
```

```python
# method overriding 
class Animal:
    def sound(self):
        return 'Animal sound'
class Dog(Animal):
    def sound(self):
        return 'Bark'
d = Dog()
d.sound()
```

```python
# method overloading
class Math:
    def add(self, a, b, c=0):
        return a + b + c
m = Math()
print(m.add(12,15))
print(m.add(12,15,18))
```

```python
class Student:
    def __init__(self):
        self._name = ""       # protected instance attribute
        self.__marks = 0

    # Getter
    @property
    def marks(self):
        return self.__marks

    # Setter
    @marks.setter
    def marks(self, value):
        if 0 < value <= 100:
            self.__marks = value
        else:
            return 'Invalid input'
    

    def grade(self):
        if self.__marks > 80:
            return 'A'
        elif self.__marks > 60:
            return 'B'
        else:
            return 'C'
```

```python
s = Student()
s.marks = 69
s.grade()
```
